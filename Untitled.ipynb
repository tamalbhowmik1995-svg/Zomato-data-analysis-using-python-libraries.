```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8')
```

```python
#display summary statistics for numerical column
df=pd.read_csv('zomato.csv', encoding='latin-1')
df.head()
```

```python
#check for missing values
print('number of rows:', df.shape[0])
print('number of column:', df.shape[1])
df.info()
```

```python

df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
#check duplicate and remove duplicate
print('Duplicate Rows:', df.duplicated().sum())
df=df.drop_duplicates()
print('After removing duplicates:', df.duplicated().sum())
```

```python
#unique countries of location
df['Country Code'].unique()
```

```python
#Most Common cuisines
top_cuisines=df['Cuisines'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index,palette='viridis')
plt.title('Top 10 most populart cuisines')
plt.xlabel('number of resturants')
plt.ylabel('cuisins type')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='orange')
plt.title('distribution of resturant rating')
plt.ylabel('Aggregate Rating')
plt.xlabel('Count')
plt.show()
```

```python
#average rating by city
avg_rating_city=df.groupby('City')['Aggregate rating'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(16,6))
sns.barplot(x=avg_rating_city.values, y=avg_rating_city.index, palette='coolwarm')
plt.title('Top 10 cities by average rating')
plt.xlabel('Average rating')
plt.ylabel('City')
plt.show()
```

```python
#checking availability of onlie delivery
delivery=df['Has Online delivery'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(delivery,labels=delivery.index,autopct='%1.1f%%',startangle=90, colors=['#FF9999','#66B2FF'])
plt.title('Online deivery availability')
plt.show()
```

```python
#resturants with and without table booking
table_booking=df['Has Table booking'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(table_booking,labels=table_booking.index,autopct='%1.1f%%',startangle=90, colors=['blue','green'],wedgeprops=dict(width=0.5))
plt.title('table booking availability')
plt.show()
```

```python
#scatter plot between number of votes and ratings
plt.figure(figsize=(10,6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, color='purple')
plt.title('votes vs rating')
plt.xlabel('number of votes')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
#correlation between numerical columns
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True,cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```
